```python
def rearrange_col(dataframe, col_to_move, reference_col, right=True):
    col_list = dataframe.columns.values.tolist()
    col_list2 = [x for x in col_list if x != col_to_move]
    reference_idx = col_list2.index(reference_col)
    if right==True:
        col_list3 = []
        for y in col_list2:
            col_list3.append(y)
            if y == reference_col:
                col_list3.append(col_to_move)
        return dataframe[col_list3]
    else:
        col_list3 = []
        for y in col_list2:
            col_list3.append(y)
            if y == reference_col:
                col_list3.insert(-2, col_to_move)
        return  dataframe[col_list3]
```

```python
def EasyCheckData(result_df, hit_N, query_df, words):
    result_dic = result_df.to_dict(orient='index')
    hit_sel = result_dic[hit_N] #HIT to check
    stim_id = hit_sel['stim_id']
    resp = hit_sel['response'].split('_')
    result2build = query_df
    result2build.drop(result2build.columns[1:], axis=1, inplace=True)
    result2build['response'] = resp
    result2build['word'] = words[stim_id-1]
    result2build = rearrange_col(result2build, 'word', 'feature')
    #print(result2build)
    return result2build
```
