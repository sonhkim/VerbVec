```python
### v4
###this code works except the Propbank read here no longer uses the 'f' tag to distinguish PP args! Bummer.

preposition = ['aboard', 'about', 'above', 'across', 'after', 'against', 'along', 'amid', 'among', 'anti', 
               'around', 'as', 'at', 'before', 'behind', 'below', 'beside', 'besides', 'between', 'beyond', 
               'by', 'concerning', 'considering', 'despite', 'down', 'during', 'except', 'excepting', 'excluding',
              'following', 'for', 'from', 'in', 'inside', 'into', 'near', 'of', 'off', 'on', 'onto', 'opposite', 
              'outside', 'over', 'per', 'regarding', 'since', 'than', 'through', 'to', 'toward', 'towards', 
              'under', 'until', 'up', 'upon', 'versus', 'via', 'with', 'within', 'without']

arg_dict = {} #dictionary containing argument configuration for all lemmas
unerg_red = []  #red means redundant
unacc_red = []
trans_red = []

def make_arg_str(es):
    arg_set = es.findall('arg')
    arg_per_example = []
    for arg in arg_set:
        if arg.text is not None: # safe for execption !!
            if (arg.get('f') == '') and (arg.text.split()[0] not in preposition):
                arg_per_example.append(arg.attrib['n'])
    arg_str = concatenate_list_data(arg_per_example)
    return arg_str


def get_key_value(es):
    infl_set = es.findall('inflection')
    if 'name' in es.keys():
        if ('passiv' in str(es.attrib['name'])) or ('-n' in str(es.attrib['name'])):
            #return es.attrib['name'], 'NameFail'
            return es.attrib['name'], ('NameFail')
        else:
            if len(infl_set)==0:
                #return es.attrib['name'], make_arg_str(es)
                return es.attrib['name'], (make_arg_str(es), es.find('text').text)
            
            elif (len(infl_set)>0) and (infl_set[0].attrib['voice'] != 'passive'):
                return es.attrib['name'], (make_arg_str(es), es.find('text').text)                          

            else:
                return es.attrib['name'], ('InflFail')

    else: #'name' not in es.keys():
        if len(infl_set)==0:
            return 'noname', (make_arg_str(es), es.find('text').text)                 

        elif (len(infl_set)>0) and (infl_set[0].attrib['voice'] != 'passive'):                        
            return 'noname', (make_arg_str(es), es.find('text').text)                                     

        else:
            return 'noname', ('InflFail')


'''step1'''
#lemma_in_propbank2 = ['throw']
for lem in lemma_in_propbank:
    tree = ET.parse('/Users/songheekim/Google Drive/Primary/Projects/VerbVector/propbank/frames/%s.xml' % (lem))
    root = tree.getroot()
    predicate_set = root.findall('predicate')

    for pred in predicate_set:
        if pred.attrib['lemma'] == lem:
            example_set = pred.findall('roleset/example')
            arg_str_per_lemma = {}
            for es in example_set:
                key, value = get_key_value(es)
                arg_str_per_lemma[key] = value
            arg_dict[lem] = arg_str_per_lemma


'''step2'''
#print (arg_dict['abandon'])
for k, v in arg_dict.items():
    for kk, vv in v.items():
#        print (vv)
#    print (k,v)
        if re.fullmatch(unacc_str, vv[0]) != None:
            unacc_red.append(k)
        elif re.fullmatch(trans_str, vv[0]) != None:
            trans_red.append(k)
        elif re.fullmatch(unerg_str, vv[0]) != None:
            unerg_red.append(k)

    
unacc = sorted(list(set(unacc_red)))
unerg = sorted(list(set(unerg_red)))
trans = sorted(list(set(trans_red)))
```
